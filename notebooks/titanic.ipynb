```python
import pandas as pd
```

```python
train_csv = pd.read_csv('../datasets/titanic/train.csv')
```

```python
train_csv.describe()
```

Extract `Survived` as y_train

```python
y_train = train_csv['Survived']
```

Drop columns from train set which are not useful for classification

```python
X_train = train_csv.drop('Survived', axis = 1)
```

```python
X_train = X_train.drop(['Name','PassengerId','Ticket','Fare','Cabin','Embarked'] ,axis =1 )
```

Replace Sex with 0 for female and 1 for male

```python
X_train['Sex'].replace(['female','male'],[0,1],inplace = True)
```

```python
median_age = X_train['Age'].median()
```

```python
X_train['Age'].fillna(median_age, inplace =  True)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import SGDClassifier
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
knn_clf  = KNeighborsClassifier()
sgd_clf = SGDClassifier(max_iter=5, tol=-np.infty,random_state=42)
forest_clf = RandomForestClassifier(random_state = 42)
```

```python
forest_score = cross_val_score(forest_clf, X_train, y_train, cv = 10, scoring = 'accuracy')
```

```python
forest_score.mean()
```

```python
test_csv = pd.read_csv('../datasets/titanic/test.csv')
```

```python
test_csv = test_csv.drop(['Name','PassengerId','Ticket','Fare','Cabin','Embarked'] ,axis =1 )
```

```python
test_csv['Age'].fillna(median_age, inplace =  True)
```

```python
test_csv['Sex'].replace(['female','male'],[0,1],inplace = True)
```

```python
forest_clf.fit( X_train, y_train)
```

```python
prediction = forest_clf.predict(test_csv)
```

```python
len(test_csv)
```

```python
pred_df = pd.DataFrame({'Survived': prediction})

```
